```python
nome = " MARIA EDUARDA ALMEIDA "
email = " MARIA.ALMEIDA@EMPRESA.COM "

#1 Remover erspaços extras
nome = nome.strip()
email = email.strip()

#2 - converter para minúsculo
nome = nome.lower()
email = email.lower()

#3 - substituir o email
email = email.replace("empresa.com","gmail.com")

#4 - Qtd de caracteres
quantidade = len(nome)

#5 - ultimo/primeiro
ultima_letra = nome[-1]
primeira_letra = nome[0]
print("nome tratado:",nome)
print("email tratado:",email)
print("quantidade de caracteres:",quantidade)
print("ultima letra:",ultima_letra)
print("primeira letra:",primeira_letra)


```
